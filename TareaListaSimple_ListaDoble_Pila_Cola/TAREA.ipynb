````python
%%writefile ListaSimple.cpp
#include <iostream>
#include <string>
#include <functional>
#include <cstdlib>

using namespace std;

template<class Carmen>
struct Nodo{
    Carmen valor;
    Nodo *sgt;
    Nodo(Carmen valor, Nodo *sgt = nullptr){
        this->valor = valor;
        this->sgt = sgt;
    }  
};

template<class Carmen>
class ListaSimple{
    Nodo<Carmen> *inicio;
    unsigned len;

public:
    ListaSimple(){
        inicio = nullptr;
        len = 0;
    }
    ~ListaSimple(){
        Nodo<Carmen> *aux = inicio;
        while(aux != nullptr){
            delete inicio;
            aux = aux->sgt;
            inicio = aux;
        }
    }

    int size(){
        return len;
    }

    void agregarInicio(Carmen va){
        Nodo<Carmen> *nuevo = new Nodo<Carmen>(va);
        if(inicio == nullptr){
            inicio = nuevo;
        }else{
            nuevo->sgt = inicio;
            inicio = nuevo;
        }
        len++;
    }

    void agregarInicio(Nodo<Carmen>* nuevo){
        if(inicio == nullptr){
            inicio = nuevo;
        }else{
            nuevo->sgt = inicio;
            inicio = nuevo;   
        }
        len++;
    }

    void agregarFinal(Carmen va){
        Nodo<Carmen> *nuevo = new Nodo<Carmen>(va);
        Nodo<Carmen> *aux = inicio;
        if(inicio == nullptr){
            inicio = nuevo;
            len++;
        }else{
            for(size_t i = 1; i < len; i++){
                aux = aux->sgt;
            }
            aux->sgt = nuevo;
            len++;
        }
    }

    void agregarFinal(Nodo<Carmen> *nuevo){
        Nodo<Carmen> *aux = inicio;
        if(inicio == nullptr){
            inicio = nuevo;
            len++;
        }else{
            for(size_t i = 1; i < len; i++){
                aux =  aux->sgt;
            }
            aux->sgt = nuevo;
        }
    }

    void agregarPos(int pos, Carmen va){
        if(pos == 1){
            agregarInicio(va);
        }
        if(pos == len + 1){
            agregarFinal(va);
        }else{
            Nodo<Carmen> *aux = inicio;
            for(size_t i = 1; i < pos - 1; i++){
                aux = aux->sgt;
            }
            Nodo<Carmen> *nuevo = new Nodo<Carmen>(va, aux->sgt);
            aux->sgt = nuevo;
            len++;
        }
    }

    void agregarPos(int pos, Nodo<Carmen> *nuevo){
        if(pos == 1){
            agregarInicio(nuevo);
        }
        if(pos == len + 1){
            agregarFinal(nuevo);
        }else{
            Nodo<Carmen> *aux = inicio;
            for(size_t i = 1; i < pos - 1; i++){
                aux = aux->sgt;
            }
            nuevo->sgt = aux->sgt;
            aux->sgt = nuevo;
            len++;
        }
    }

    void imprimir(){
        Nodo<Carmen> *aux = inicio;
        while(aux != nullptr){
            cout << aux->valor << " ";
            aux = aux->sgt;
        }
    }

    void imprimir(function<void(Carmen)> fMostrar){
        Nodo<Carmen> *aux = inicio;
        for(size_t i = 0; i < len; i++){
            fMostrar(aux->valor);
            cout << endl;

            aux =  aux->sgt;
        }
    }


    void eliminarPrimero(){
        if(inicio != nullptr){
            Nodo<Carmen> *aux = inicio;
            inicio = inicio->sgt;
            delete aux;
            len--;
        }
    }

    void eliminarFinal(){
        Nodo<Carmen> *aux = inicio;
        if(inicio != nullptr){
            for(size_t i = 1; i < len - 1; i++){
                aux = aux->sgt;
            }
            Nodo<Carmen> *aux2 = aux->sgt;
            aux->sgt = nullptr;
            delete aux2;
            len--;
       }
    }

    void eliminarPos(int pos){
        Nodo<Carmen> *aux = inicio;
        if(pos == 1){
            eliminarPrimero();
        }
        if(pos == len){
            eliminarFinal();
        }else{
            for(size_t i = 1; i < pos - 1; i++){
                aux = aux->sgt;
            }
            Nodo<Carmen> *aux2 = aux->sgt;
            aux->sgt = aux2->sgt;
            delete aux2;
            len--;
        }
    }

    int cantidad(){
        int contador = 0;
        Nodo<Carmen>* aux = inicio;
        while(aux != nullptr){
            contador++;
            aux = aux->sgt;
        }
        return contador;
    }

    bool recuperarPos(Carmen nuevo, function <bool(Carmen, Carmen)> comp){
        Nodo<Carmen> *aux = inicio;
        int contador = 0;
        while(aux != nullptr){
            contador++;
            if(comp(nuevo, aux->valor)){
                cout << "Valor encontrado, en la pos " << contador << endl;
                return true;
                break;
            }
            aux = aux->sgt;
        }
        return false;
    }
};

class Virus{
    string nombre;
    short idx;

public:
    Virus(string nombre, short idx){
        this->nombre = nombre;
        this->idx = idx;
    }
    string toString(){return nombre + " " + to_string(idx);}
    string getNombre(){return nombre;}

    short getIdx(){return idx;}


};

void prueba1(){
    ListaSimple<Virus*> ls;
    ls.agregarInicio(new Virus("Covid", 1));
    ls.agregarInicio(new Virus("Influenza", 2));
    ls.agregarInicio(new Virus("Colera", 3));
    ls.agregarPos(2,new Virus("Ebola", 4));
    

    cout << ls.cantidad() << endl;
    ls.imprimir([](Virus *v){cout << v->toString();});


    ls.eliminarFinal();
    cout << endl;
    ls.imprimir([](Virus *v){cout << v->toString();});

    cout << ls.recuperarPos(new Virus("Influenza", 2), [](Virus* x, Virus* y){return x->getNombre() == y->getNombre()  && x->getIdx() == y->getIdx();}) << endl;
}

int main(){    
    prueba1();
    return 0;
}
````

````python
%%writefile Cola.cpp
#include <iostream>
#include <functional>
#define NULO 0

using namespace std;

template<class Carmen>
struct Nodo{
    Carmen valor;
    Nodo *sgt;
    Nodo *atras;
    Nodo(Carmen valor, Nodo *sgt = nullptr, Nodo *atras = nullptr){
        this->valor = valor;
        this->sgt = sgt;
        this->atras = atras;
    }
};

template<class Carmen>
class Cola{
      Nodo<Carmen> *inicio;
      Nodo<Carmen> *fin;
      unsigned int len;

public:
      Cola(){
          inicio = nullptr;
          fin = nullptr;
          len = 0;
      }

      ~Cola(){
          Nodo<Carmen> *aux = inicio;
          while(aux != nullptr){
              Nodo<Carmen> *aux2 = aux;
              delete aux2;
              aux = aux->sgt;
          }
      }

      void push(Carmen va){
          Nodo<Carmen> *nuevo = new Nodo<Carmen>(va);
          if(inicio == nullptr){
              inicio = fin = nuevo;
          }else{
              fin->sgt = nuevo;
              nuevo->atras = fin;
              fin = nuevo;
          }
          len++;
      }

      void pop(){
          if(inicio == nullptr){
              return;
          }
          Nodo<Carmen> *aux = inicio;
          if(len == 1){
              inicio = fin = nullptr;
          }else{
              inicio = inicio->sgt;
              inicio->atras = nullptr;
          }
          
          delete aux;
          len--;
      }

      Carmen top(){
          if(len > 0){
              return inicio->valor;
          }else{
              return NULO;
          }
      }

      void top(function<void(Carmen)> fTop){
        if(len > 0){
            fTop(inicio->valor);
        }
    }
};
class Virus{
  string nombre;
  short idx;
  public:
  Virus(string nombre,short idx):nombre(nombre),idx(idx){}
  string toString(){return nombre+" "+to_string(idx)+"\n";}
  string getNombre(){return nombre;}
};

void ejemplo2(){
    Cola<Virus*> ls;   
    ls.push(new Virus("Covid",1));
    ls.push(new Virus("Influenza",2));
    ls.push(new Virus("Colera",3));     

    ls.top([](Virus* v){cout<<v->toString();});    
    ls.pop();
    ls.top([](Virus* v){cout<<v->toString();});    
    ls.pop();
    ls.top([](Virus* v){cout<<v->toString();});    
    ls.pop();
    ls.top([](Virus* v){cout<<v->toString();});    
    ls.pop();
     

}
int main(){
    ejemplo2();
}
````
