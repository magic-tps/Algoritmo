```python
%%writefile ListaSimpleCircular.cpp
#include <iostream>

using namespace std;

template<class Carmen>
class ListaSimpleCircular{
    struct Nodo{
          Carmen valor;
          Nodo *sgt;
          Nodo(Carmen valor, Nodo *sgt = nullptr){
              this->valor = valor;
              this->sgt = sgt;
          }
    };
    Nodo *inicio;
    Nodo *final;
    unsigned int len;

public:

ListaSimpleCircular(){
    len = 0;
    inicio = nullptr;
    final = nullptr;
}

int size(){
    return len;
}
void agregarInicio(Carmen va){
    Nodo *nuevo = new Nodo(va);
    if(inicio ==  nullptr){
        inicio = nuevo;
        final = inicio;
        len++;
    }else if(len == 1){
        nuevo->sgt = inicio;
        inicio = nuevo;
        inicio->sgt = final;
        final->sgt = inicio;
        len++;
    }else{
        nuevo->sgt = inicio;
        inicio = nuevo;
        final->sgt = inicio;
        len++;
    }
}

void agregarPos(int pos, Carmen va){
    if(pos == 1){
        agregarInicio(va);
    }else{
        Nodo* aux = inicio;
        for(size_t i = 1; i < pos - 1; i++){
            aux = aux->sgt;
        }
        Nodo *nuevo = new Nodo(va, aux->sgt);
        aux->sgt = nuevo;

        Nodo*aux2 = inicio;
        for(size_t  i = 0; i < len; i++){
            aux2 = aux2->sgt;
        }
        final  = aux2;
        final->sgt = inicio;

        len++;
    }
}

void agregarFinal(Carmen va){
    agregarPos(len + 1, va);
}

void eliminarInicio(){
    if(inicio != nullptr){
        Nodo *aux = inicio;
        inicio = inicio->sgt;
        delete aux;
        final->sgt = inicio;
       
    }
}

void eliminarPos(int pos){
    if(pos == 1){
        eliminarInicio();
    }
    if(pos == len){
        eliminarFinal();
    }else{
        Nodo *aux = inicio;
        for(int i = 1; i < pos - 1; i++){
            aux = aux->sgt;
        }
        Nodo *aux2 = aux->sgt;
        aux->sgt = aux2->sgt;
        delete aux2;
        len--;
    }
}


//////////////////////////////Falla en algo u.u
void eliminarFinal(){
    Nodo *aux = inicio;
    for(int i = 1; i < len - 1; i++){
        aux = aux->sgt;
    }
    Nodo *aux2 = aux->sgt;
    final = aux;
    final->sgt = inicio;
    //delete aux2;
}

void imprimir(){
    Nodo *aux = inicio;
    int contador = 0;
    if(inicio != nullptr){
        do{
           contador++;
           cout << aux->valor << " ";
           aux = aux->sgt; 
        }while(aux != inicio);
    }
    cout << endl << contador;
}

};

int main(){
    ListaSimpleCircular<int> lista;
    lista.agregarInicio(9);
    lista.agregarInicio(8);
    lista.agregarInicio(7);
    lista.agregarInicio(6);
    lista.agregarInicio(5);
    lista.agregarInicio(4);
    lista.agregarInicio(3);

    cout << endl;
    lista.imprimir();
    cout << endl;

    lista.agregarPos(8, 40);
    cout << endl;
    lista.imprimir();
    cout << endl;

    lista.agregarFinal(50);
    cout << endl;
    lista.imprimir();
    cout << endl;

    lista.eliminarInicio();
    cout << endl;
    lista.imprimir();
    cout << endl;

    lista.eliminarPos(3);
    cout << endl;
    lista.imprimir();
    cout << endl;

    lista.eliminarFinal();
    cout << endl;
    lista.imprimir();
    cout << endl;

    return 0;
}
```
