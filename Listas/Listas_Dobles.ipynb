```python
%%writefile ListaDoble.cpp
#include <iostream>

using namespace std;

template<class Carmen>
class ListaDoble{
    struct Nodo{
        Carmen valor;
        Nodo *sgt;
        Nodo *atras;

        Nodo(Carmen valor, Nodo *sgt = nullptr, Nodo *atras = nullptr){
            this->valor = valor;
            this->sgt = sgt;
            this->atras = atras;
        }
    };

    unsigned int len;
    Nodo *inicio;

public:

ListaDoble(){
    len = 0;
    inicio = nullptr;
}

~ListaDoble(){}

int size(){
    return len;
}


void agregarInicio(Carmen va){
    Nodo *nuevo = new Nodo(va);
    if(inicio == nullptr){
        inicio = nuevo;
        len++;
    }else{
        nuevo->sgt = inicio;
        inicio->atras = nuevo;
        inicio = nuevo;
        inicio->atras = nullptr;
        len++;
    }
}

void agregarPos(int pos, Carmen va){
    Nodo* aux = inicio;
    if(pos == 1){
        agregarInicio(va);
    }else{
        for(size_t i = 1; i < pos - 1; i++){
            aux = aux->sgt;
        }
        Nodo* nuevo = new Nodo(va, aux->sgt, aux);
        aux->sgt = nuevo;
        len++;
    }
}

void agregarFinal(Carmen va){
    if(inicio == nullptr){
        agregarInicio(va);
    }else{
        agregarPos(len + 1, va);
    }
}

void eliminarInicio(){
    if(inicio != nullptr){
        Nodo *aux =  inicio;
        inicio = inicio->sgt;
        inicio->atras = nullptr;
        delete aux;
        len--;
    }
}

void eliminarPos(int pos){
    if(pos == 1){
        eliminarInicio();
    }
    if(pos == len){
        eliminarFinal();
    }else{
        Nodo *aux = inicio;
        for(size_t i = 1; i < pos - 1; i++){
            aux = aux->sgt;
        }
        Nodo *aux2 = aux->sgt;
        aux->sgt = aux2->sgt;
        aux2->sgt->atras = aux;
        delete aux2;
        len--;
    }
  
}

void eliminarFinal(){
    if(inicio != nullptr){
        Nodo *aux = inicio;
        for(size_t i = 1; i < len - 1; i++){
            aux = aux->sgt;
        }
        Nodo*aux2 = aux->sgt;
        aux->sgt = nullptr;
        delete aux2;
        len--;
    }
    
}

void buscarElemento(Carmen va){
    Nodo *aux = inicio;
    int contador = 0;
    while(aux != nullptr){
        contador++;
        if(aux->valor = va){
            cout << "Se encontro el valor  " << va << " y se encuentra en la pos " << contador << endl;
            break;
        }
        aux = aux->sgt;
    }
}

void imprimir(){
    Nodo *aux = inicio;
    int contador = 0;
    while(aux != nullptr){
        cout << aux->valor << " ";
        aux = aux->sgt;
        contador++;
    }
    cout << endl;
    cout << contador;
}




};

int main(){
    ListaDoble<int> lista;
    lista.agregarInicio(9);
    lista.agregarInicio(8);
    lista.agregarInicio(7);
    lista.agregarInicio(6);
    lista.agregarInicio(5);
    lista.agregarInicio(4);
    lista.agregarInicio(3);
    lista.agregarPos(2, 10);
    lista.agregarFinal(15);
    
    lista.imprimir();
    
    cout << endl;
    lista.eliminarInicio();
    lista.imprimir();
    cout << endl;

    cout << endl;
    lista.eliminarPos(lista.size());
    lista.imprimir();

    cout << endl;
    lista.eliminarFinal();
    lista.imprimir();

    cout << endl;
    lista.buscarElemento(5);
    lista.imprimir();

    return 0;
}
```
