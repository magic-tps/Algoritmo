```python
%%writefile ListaSimple.cpp
#include <iostream>
 
using namespace std;
 
template<class Carmen>
class Lista{
    struct Nodo{
        Carmen valor;
        Nodo* sgt;
        Nodo(Carmen valor, Nodo* sgt = nullptr){
            this->valor =  valor;
            this->sgt = sgt;
        }
    };
    Nodo* inicio;
    unsigned int len;
public:
 
Lista(){
    inicio = nullptr;
    len = 0;
}
 
int size(){
    return len;
}
 
void agregarInicio(Carmen va){
    Nodo *nuevo = new Nodo(va);
    if(inicio == nullptr){
        inicio = nuevo;
        len++;
    }else{
        nuevo->sgt = inicio;
        inicio = nuevo;
        len++;
    }
}
 
void agregarPos(Carmen va, int pos){
    if(pos == 0){
        agregarInicio(va);
    }else{
        Nodo* aux = inicio;
        for(int i = 1; i < pos - 1; i++){
            aux = aux->sgt;
        } 
        Nodo* nuevo = new Nodo(va, aux->sgt);
        aux->sgt = nuevo;
        len++;
    }
}
 
void agregarFinal(Carmen va){
    if(inicio == nullptr){
        agregarInicio(va);
    }else{
        agregarPos(va, len + 1);
    }
}
 
 
void imprimirLista(){
    Nodo* aux =  inicio;
    while(aux != nullptr){
        cout << aux->valor << " ";
        aux = aux->sgt;
    }
}
 
void eliminarInicio(){
    if(inicio != nullptr){
        Nodo*aux = inicio;
        inicio = inicio->sgt;
        delete aux;
        len--;
    }
 
}
 
void eliminarPos(int pos){
    if(pos == 0){
        eliminarInicio();
    }else{
        Nodo* aux = inicio;
        for(int i = 1; i < pos - 1; ++i){
            aux = aux->sgt;
        }
        Nodo*aux2 = aux->sgt;
        aux->sgt = aux2->sgt;
        delete aux2;
        len--;
    }
}
 
void eliminarFinal(){
    if(inicio != nullptr){
        eliminarPos(len);
    }
}
 
void buscarElemento(Carmen va){
    Nodo*aux = inicio;
    while(aux != nullptr){
        if(aux->valor == va){
            cout << "Elemento encontrado y es el " <<  va;
            break;
        }
        aux = aux->sgt;
    }
}
 
void buscarPos(Carmen va){
    Nodo* aux = inicio;
    int i = 0;
    while(aux != nullptr){
         if(aux->valor == va){
            cout << "La posicion es: " << i + 1;
            break;
        }   
        i++;
        aux = aux->sgt;
    }
}
 
/////////////////////////// NO SALE
void rotarLista(){
    Nodo *nuevoInicio;
    Nodo *aux = inicio;
    int auxlen = len;
    cout << "Hola";
    while(true){
 
        nuevoInicio->sgt = aux;
        aux = aux->sgt;
        auxlen--;
        if(auxlen == 0){break;}
    }
    Nodo *aux2 = nuevoInicio->sgt;
    while(aux2 != nullptr){
        cout << aux2->valor;
        aux2 = aux2->sgt;
    }
}
///////////////////////////


void editarInicio(Carmen va){
    if(inicio != nullptr){
        inicio->valor = va;
    }
}

void editarPos(int pos, Carmen va){
    Nodo *aux = inicio;
    if(pos <= 0 && pos > len){return;}
    if(pos == 1){
        editarInicio(va);
    }else{
        for(size_t i = 1; i < pos; i++){
            aux = aux->sgt;
        }
        aux->valor = va;
    }
}

void editarFinal(Carmen va){
    editarPos(len, va);
}
 
};
 
/*
friend ostream& operator << (ostream o Clase & c){
    o << c.... << " " << c...  << endl;
    return o;  
}
*/
 
int main(){
    Lista<int> lista;
    lista.agregarInicio(5);
    lista.agregarInicio(6);
    lista.agregarInicio(8);
    lista.agregarInicio(9);
    lista.agregarPos(7, 3);
    lista.agregarFinal(10);
    cout << lista.size() << endl;
    lista.imprimirLista();

    cout << endl;
    lista.editarInicio(100);
    lista.imprimirLista();
    
    cout << endl;
    lista.editarPos(2, 50);
    lista.imprimirLista();

    cout << endl;
    lista.editarFinal(50);
    lista.imprimirLista();
 
    return 0;
}
```
